```python
#linear search

arr=[12,35,6,3,5,8,33,98,4,11,56]
n=len(arr)
key=int(input("enter the key to search"))

def linearsearch(arr,n,key):
    for i in range(0,n):
        if arr[i]==key:
            print("key found at index ",i)
            return 
    return "key not found in array"

linearsearch(arr,n,key)
```

```python
#binary search
arr=[12,35,6,3,5,8,33,98,4,11,56]
arr.sort()
print("the sorted array is ",arr)
n=len(arr)
key=int(input("enter the key to search"))

def binarysearch(arr,low,high,key):
    if(low>=high):
        print("key not found")
        return
    
    mid=(low+high)//2

    if arr[mid]==key:
        print("key found at index ",mid)
        return

    if key<arr[mid]:
        binarysearch(arr,low,mid,key)
    elif(key>mid):
        binarysearch(arr,mid+1,high,key)

binarysearch(arr,0,n-1,key)
```
