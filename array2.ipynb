```python
#longest sbaray with k
#Method1---------

arr=[1,2,3,1,1,1,2,4,2,3]
n=len(arr)
l=0
k=4
for i in range(n):
    sum=0
    for j in range(i,n):
        sum+=arr[0]
        if sum==k:
            l=max(l,j-i+1)

print(l)
```

```python
#Method --------2
arr=[1,2,3,1,1,1,2,4,2,3]
n=len(arr)
k=4
left,right=0,0
sum=arr[0]
maxl=0
while(right<n):
    while(left<=right and sum>k):
        sum-=arr[left]
        left+=1

    if sum==k:
        maxl=max(maxl,right-left+1)
        
    right+=1

    if(right<n):
        sum+=arr[right]
    
print(maxl)

```

```python
#Two sum problem 
#Methd 1
arr=[2,6,5,8,11]
target=14
n=len(arr)
for i in range(n):
    for j in range(i+1,n):
        if arr[i]+arr[j]==target:
            print("firs value",arr[i]," second value", arr[j])

```

```python
#Method 2

arr=[2,6,5,8,11]
target=14
n=len(arr)
left=0
right=n-1

while(left<right):
    sum=arr[left]+arr[right]
    if sum==target:
        print(left, right)
        break

    elif(sum<target):
        left+=1

    else:
        right-=1
        
```

```python
#Sort an array of 0,1,2

#Method 1

arr=[0,1,2,0,1,2,1,2,0,0,0,1]
n=len(arr)
count0=0
count1=0
count2=0
for i in range(n):
    if arr[i]==0:count0+=1
    if arr[i]==1:count1+=1
    else:count2+=1

for i in range(0,count0):arr[i]=0
for i in range(count0,count0+count1):arr[i]=1
for i in range(count0+count1,n):arr[i]=2

print(arr)
```

```python
#Method 2

#Dutch National Flag Algoritm

arr=[0,1,2,0,1,2,1,2,0,0,0,1]
n=len(arr)
low=0
mid=0
high=n-1
while(mid<=high):
    if arr[mid]==0:
        arr[low],arr[mid]=arr[mid],arr[low]
        low+=1
        mid+=1
    elif arr[mid]==1:
        mid+=1
    elif arr[mid]==2:
        arr[mid],arr[high]=arr[high],arr[mid]
        high-=1

print(arr)
```
