```python
arr=[23,35,122,133,4,1223,7,86,14,10,9770]
large=arr[0]
for i in range(len(arr)):
    if (arr[i]>large):
        large=arr[i]
print(large)
```

```python
#second largest-------------brute force
largest=arr[0]
for i in range(len(arr)):
    if (arr[i]>largest):
        largest=arr[i]
slargest=-1
for i in range(len(arr)):
    if (arr[i]>slargest and arr[i]!=largest):
        slargest=arr[i]
print(slargest)
```

```python
#second largest -------------optimal approach
l=arr[0]
sl=-1
for i in range(len(arr)):
    if(arr[i]>l):
        sl=l
        l=arr[i]
    elif(arr[i]!=l and arr[i]>sl):
        sl=arr[i]
print(l)
print(sl)
```

```python
#second smallest-----------------optimal approach
s=arr[0]
ss=float('inf')
for i in range(len(arr)):
    if(arr[i]<s):
        ss=s
        s=arr[i]
    elif(arr[i]!=s and arr[i]<ss):
        ss=arr[i]
print(s)
print(ss)
```

```python
#remove duplicate from array 
arr=[1,2,3,4,4,4,5,5,6]
for i in range(len(arr)-1):
    for j in range(len(arr)-1,i,-1):
        if arr[i]==arr[j]:
            arr.remove(arr[j])
print(arr)
```

```python
#left rotate by 1 place
arr=[1,2,3,4,5,6,7,8]
n=len(arr)
temp=arr[0]
for i in range(1,n):
    arr[i-1]=arr[i]
arr[n-1]=temp
print(arr)
```

```python
#left rotate by d places brute appraoch

arr=[1,2,3,4,5,6,7,8]
n=len(arr)
temp=[]
d=int(input("enter the d value "))
d=d%n
print("the d places to rotate value is", d)
for i in range(0,d):
    temp.append(arr[i])

for i in range(d,n):
    arr[i-d]=arr[i]

for i in range(n-d,n):
    arr[i]=temp[i-(n-d)]

print(arr)
```

```python
#left rotae by d optimal approach
arr=[1,2,3,4,5,6,7,8]
n=len(arr)
d=int(input("enter the d value "))

def reversearray(arr,low,high):
    start=low
    end=high
    while(start<=end):
        arr[start],arr[end]=arr[end],arr[start]
        start+=1
        end-=1

reversearray(arr,0,d-1)
print(arr)

reversearray(arr,d,n-1)
print(arr)

reversearray(arr,0,n-1)
print(arr)

```

```python
#right rotate by 1 place
arr=[1,2,3,4,5,6,7,8]
n=len(arr)

temp=arr[n-1]

for i in range(n-1,0,-1):
    arr[i]=arr[i-1]
arr[0]=temp
print(arr)

```

```python
#right rotate by d place

arr=[1,2,3,4,5,6,7,8]
n=len(arr)
d=3
temp=[]
for i in range(n-d,n):
    temp.append(arr[i])
print(temp)

for i in range(n-1,0,-1):
    arr[i]=arr[(n-d)-i]

print(arr)

for i in range(0,d):
    arr[i]=temp[i]
print(arr)
```

```python
#move all zeros to the end of array
#brute

arr=[1,0,2,3,2,0,0,4,5,3]
n=len(arr)
temp=[]
for i in range(0,n):
    if arr[i]!=0:
        temp.append(arr[i])

nonZero=len(temp)

for i in range(0,nonZero):
    arr[i]=temp[i]

for i in range(nonZero,n):
    arr[i]=0

print("the modified array is ",arr)
```

```python
#move all zeros to the end of array
#optimal
arr=[1,0,2,3,2,0,0,4,5,3]
n=len(arr)

j=-1
for i in range(0,n):
    if arr[i]==0:
        j=i
        break

for i in range(j+1,n):
    if arr[i]!=0:
        arr[i],arr[j]=arr[j],arr[i]
        j+=1


print("the modified array is ",arr)
```
