```python
#bubble sort
arr=[25,9,35,3,80,45,21,4,20,40]
n=len(arr)
def bubblesort(arr,n):
    for i in range(n):
        for j in range(n-i-1):
            if arr[j]>arr[j+1]:
                arr[j],arr[j+1]=arr[j+1],arr[j]
    return arr
result=bubblesort(arr,n)
print(result)
```

```python
arr=[25,9,35,3,80,45,21,4,20,40]
n=len(arr)
def insertionsort(arr,n):
    for i in range(n):
        j=i
        while(j>0 and arr[j-1]>arr[j]):
            arr[j-1],arr[j]=arr[j],arr[j-1]
            j-=1
    return arr


result=insertionsort(arr,n)
print(result)
```

```python
arr=[25,9,35,3,80,45,21,4,20,40]
n=len(arr)
def selectionsort(arr,n):
    for i in range(n-1):
        mini=i
        for j in range(i,n):
            if arr[j]<arr[mini]:
                mini=j
        arr[mini],arr[i]=arr[i],arr[mini]
    return arr
result=selectionsort(arr,n)
print(result)
```

```python
#mergesort
arr=[25,19,350,3,180,45,21,4,20,40]
n=len(arr)


def mergesort(arr,low,high):
    if(low>=high):
        return
    mid=(low+high)//2
    mergesort(arr,low,mid)
    mergesort(arr,mid+1,high)
    merge(arr,low,mid,high)


def merge(arr,low,mid,high):
    left=low
    right=mid+1
    temp=[]
    while(left<=mid and right<=high):
        if arr[left]<=arr[right]:
            temp.append(arr[left])
            left+=1
        else:
            temp.append(arr[right])
            right+=1
    while(left<=mid):
        temp.append(arr[left])
        left+=1
    while(right<=high):
        temp.append(arr[right])
        right+=1
    for i in range(len(temp)):
        arr[low+i]=temp[i]

mergesort(arr,0,n-1)
print(arr)

```

```python
#quicksort
arr=[30,45,24,556,21,3,2,66,2,12,33]
n=int(len(arr))

def quicksort(arr,low,high):
    if(low>=high):return
    partition=findpivot(arr,low,high)
    quicksort(arr,low,partition-1)
    quicksort(arr,partition+1,high)
def findpivot(arr,low,high):
    pivot=arr[low]
    i=low
    j=high
    while i<j:
        while(i<=high and arr[i]<=pivot):
            i+=1
        while(j>=low and arr[j]>pivot):
            j-=1
        if(i<j):
            arr[i],arr[j]=arr[j],arr[i]
    arr[low],arr[j]=arr[j],arr[low]
    return j
    

quicksort(arr,0,n-1)
print(arr)
```

```python
arr=[20,3,64,12,2,7,85,34,35,1]
n=len(arr)
key=int(input("enter the key to find"))

def linearsearch(arr,n,key):
    for i in range(n):
        if arr[i]==key:
            print("key found at ",i)
            return
    print("Key not found")
linearsearch(arr,n,key)
```

```python
arr=[20,3,64,12,2,7,85,34,35,1]
arr.sort()
n=len(arr)
key=int(input("enter the key to find"))

def binarysearch(arr,low,high,key):
    if(low>high):
        print("key not found")
        
        return

    mid=(low+high)//2

    if(arr[mid]==key):
        print("key value ",arr[mid]," found at index ",mid)
        return
    elif(key<arr[mid]):
        binarysearch(arr,low,mid-1,key)
    else:
        binarysearch(arr,mid+1,high,key)
    
binarysearch(arr,0,n-1,key)
```
